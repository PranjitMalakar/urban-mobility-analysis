```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


print("Loading dataset...")
df = pd.read_csv("urban_mobility_data.csv")
print("Dataset loaded successfully.")


df['start_time'] = pd.to_datetime(df['start_time'])


df['hour'] = df['start_time'].dt.hour
df['day'] = df['start_time'].dt.day_name()


initial_shape = df.shape
df = df[(df['duration_min'] > 0) & (df['duration_min'] < 180)]
df = df[(df['distance_km'] > 0) & (df['distance_km'] < 100)]
cleaned_shape = df.shape
print(f"Data cleaned: {initial_shape[0]} â†’ {cleaned_shape[0]} records.")


df['speed_kmh'] = df['distance_km'] / (df['duration_min'] / 60)


print("\nSummary statistics:")
print(df[['duration_min', 'distance_km', 'speed_kmh']].describe())


print("\nMissing values in dataset:")
print(df.isnull().sum())



plt.figure(figsize=(10, 5))
df['hour'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Trips by Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 5))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day'].value_counts().reindex(day_order).plot(kind='bar', color='coral')
plt.title("Trips by Day of the Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Trips")
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 6))
df['mode'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Transport Mode Share")
plt.ylabel("")
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
df.groupby('mode')['speed_kmh'].mean().sort_values().plot(kind='barh', color='mediumseagreen')
plt.title("Average Speed by Mode")
plt.xlabel("Speed (km/h)")
plt.grid(axis='x')
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
df.boxplot(column='duration_min', by='mode', grid=False)
plt.title("Trip Duration by Mode")
plt.suptitle("")
plt.xlabel("Mode of Transport")
plt.ylabel("Duration (minutes)")
plt.tight_layout()
plt.show()

print("\nUrban Mobility Pattern Analysis completed successfully.")
```
